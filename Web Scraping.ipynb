```python
import requests
import bs4

website = "http://ww2.amstat.org/publications/jse/jse_data_archive.htm"
server_response = requests.get( website )
server_response
```

```python
soup = bs4.BeautifulSoup( server_response.text, "lxml" )
link_info = []

for tag in soup.find_all( 'a' ) :
    
    link = tag['href']
    name = tag.text
    
    if name[-3:] == 'txt' :
        
        link_info.append( { 'link' : link, 'name' : name } )

link_info
```

```python
host = "http://www.amstat.org/publications/jse/"

for dataset in link_info[:3] :
    
    url = host + dataset['link']
    data_response = requests.get( url )
    
    if data_response.text[:5] == 'NAME:' :
        
        dataset['type'] = 'doc'
        
    else:
        
        dataset['type'] = 'dat'
        
print( link_info[:3] )
```

```python
def download_to_txt( file_name, data ) :
    """Saves the contents of data to a file called file_name str,
        requests.obj -> None"""
    with open( file_name, 'w' ) as txtfile :
        
        txtfile.writelines( data )
        
def strip_extension( file_name ) :
        """Takes in a file, and returns the name of the file 
            without the extension, str -> str"""
        i = -1
        
        while i > -len(file_name) :
            
            if file_name[i] == '.' :
              
                break
            
            else :
        
                i -= -1
        
        return file_name[:i]
            
testfile = "mytestfile.txt"
strip_extension( testfile )
```

```python
for dataset in link_info[:3] :
    
    url = host + dataset['link']
    data_response = requests.get( url )
    
    description = strip_extension( dataset['name'] )
    filename = description + '_' + dataset['type'] + '.txt'
    
    download_to_txt( filename, data_response.text ) 
```
