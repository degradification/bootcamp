```python
#Computer Usage Data into memory
data = pd.read_csv( "ntia_computer_usage_2015.csv" )
print( data[:5], )
```

```python
ma_count = data['MACount']
print( ma_count[:5] )

value_ma = pd.isnull( ma_count )
print( value_ma[:5] )
print( value_ma.sum ( ) )

mean_ma = ma_count.mean( )
print( "Mean of the total MA count is", mean_ma )

var_ma = ma_count.var( )
print( "Variance of the total MA count is", var_ma )

std_ma = ma_count.std( )
print( "Standard deviation of the total MA count is", std_ma )
```

```python
us_count = data['usCount']
print( us_count[:5] )

value_us = pd.isnull( us_count )
print( value_us[:5] )
print( value_us.sum ( ) )

mean_us = us_count.mean( )
print( "Mean of the total US count is", mean_us )

var_us = us_count.var( )
print( "Variance of the total US count is", var_us )

std_us = us_count.std( )
print( "Standard deviation of the total US count is", std_us )
```

```python
race_black = data['raceBlackCount']
print( race_black[:5] )

value_race = pd.isnull( race_black )
print( value_race[:5] )
print( value_race.sum ( ) )

mean_race = race_black.mean( )
print( "Mean of the total black count is", mean_race )

var_race = race_black.var( )
print( "Variance of the total black count is", var_race )

std_race = race_black.std( )
print( "Standard deviation of the total black count is", std_race )
```

```python
#Working with matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

#Creating a figure
x = np.linspace( 0, 10, 100 )

plt.style.use( 'classic' )
fig1 = plt.figure( )
plt.plot( x, np.sin( x ) )
plt.plot( x, np.cos( x ) )
plt.show( )
fig1.savefig( 'myfig.png' )
         
```

```python
fig2 =  plt.figure( )
plt.plot( x, np.sin( x ) )
plt.title( "A Sine Curve")
plt.xlabel( "x")
plt.ylabel( "sin(x)" )
plt.show( )
```

```python
#Plotting a hisogram
data = np.random.randn( 1000 )
fig3 = plt.figure( )
plt.hist( data )
plt.hist( data, bins = 30 )
plt.title( "Histogram of X")
plt.xlabel( "x")
plt.ylabel( "Frequency" )
fig3.savefig( 'myhist.png' )
plt.show( )
```

```python
#Plotting the normal distribution
norm = np.random.standard_normal( 1000 )
fig4 = plt.figure( )
plt.hist( norm, bins = 50 )
plt.title( "Standard Normal Distribution")
plt.xlabel( "x")
plt.ylabel( "Frequency" )
fig4.savefig( 'normal.jpg' )
plt.show( )
```
